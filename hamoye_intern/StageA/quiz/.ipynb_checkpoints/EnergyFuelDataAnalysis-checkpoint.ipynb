```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv")
df.head()
```

```python
df.report_year.unique()
```

```python
df.utility_id_ferc1.unique()
```

```python
len(df.record_id.unique())
```

```python
df['fuel_cost_per_unit_mmbtu'] = df['fuel_mmbtu_per_unit'] * df['fuel_cost_per_mmbtu']
df['energy_amount'] = df['fuel_qty_burned'] * df['fuel_mmbtu_per_unit']
df['fuel_delivery_cost'] = df['fuel_cost_per_unit_delivered'] * df['fuel_qty_burned']
df['fuel_burnt_cost'] = df['fuel_cost_per_unit_burned'] * df['fuel_qty_burned']
df['fuel_energy_cost'] = df['fuel_cost_per_unit_mmbtu'] * df['fuel_qty_burned'] 
```

```python
df_dict = {"year":[]}
cols = df['fuel_type_code_pudl'].unique()
for col in cols:
  df_dict["year"].append(col)
for year in df.report_year.unique():
  df_dict[year] = []
  num_list = df[df.report_year.values == year]['fuel_type_code_pudl'].value_counts().tolist()
  while len(num_list) < 6:
      num_list.append(0)
  for num in num_list:
    if year in df_dict:
      df_dict[year].append(num)

fuel_type_df = pd.DataFrame(df_dict)
fuel_type_df.set_index("year", inplace = True)
fuel_type_df = fuel_type_df.T
fuel_type_df
```

```python
sns.boxplot(y='utility_id_ferc1', x='fuel_type_code_pudl', data=df)
```

```python
sns.boxplot(x='report_year', y='fuel_type_code_pudl', data=df)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(ncols=3,nrows=2)
plt.gcf().set_size_inches(18,6, forward=False)

sns.lineplot('report_year', 'fuel_energy_cost',
             data=df, hue='fuel_type_code_pudl', 
             ax=axs[0,0]).set_title("Fuel Energy Cost")
sns.lineplot('report_year', 'fuel_burnt_cost',
             data=df, hue='fuel_type_code_pudl', 
             ax=axs[0,1]).set_title("Fuel Burning Cost")
sns.lineplot('report_year', 'fuel_delivery_cost',
             data=df, hue='fuel_type_code_pudl', 
             ax=axs[0,2]).set_title("Fuel Delivery Cost")
sns.lineplot('report_year', 'fuel_cost_per_unit_mmbtu',
             data=df, hue='fuel_type_code_pudl', 
             ax=axs[1,0]).set_title("Fuel Cost per unit for Energy")
sns.lineplot('report_year', 'fuel_cost_per_unit_burned',
             data=df, hue='fuel_type_code_pudl', 
             ax=axs[1,1]).set_title("Fuel Burning Cost per unit")
sns.lineplot('report_year', 'fuel_cost_per_unit_delivered',
             data=df, hue='fuel_type_code_pudl', 
             ax=axs[1,2]).set_title("Fuel Delivery Cost per unit")
```

```python
fuel_burnt_dict = {"year":[]}

cols = df['fuel_type_code_pudl'].unique()
for col in cols:
  fuel_burnt_dict["year"].append(col)
for year in df.report_year.unique():
  fuel_burnt_dict[year] = []
  number_list = [df[(df.report_year == year) & (df.fuel_type_code_pudl.str.contains(col))]['fuel_qty_burned'].sum() 
                  for col in cols]
  while len(number_list) < 6:
      number_list.append(0)

  for num in number_list:
    if year in df_dict:
      fuel_burnt_dict[year].append(num)

fuel_burnt_df = pd.DataFrame(fuel_burnt_dict)
fuel_burnt_df.set_index("year", inplace = True)
fuel_burnt_df = fuel_burnt_df.T
fuel_burnt_df
```

```python
sns.lineplot(x='report_year', y='energy_amount', 
             hue='fuel_type_code_pudl', 
             data=df[df.fuel_type_code_pudl != 'other']
            ).set_title("Avg Energy produced over time")
```

**Observation**: We can observed that the amount of energy produced was quite consistent for all fuel type except for the unexpected spike of nuclear energy between year 1995 and 2000.
```python
sns.lineplot(x='report_year', y='fuel_qty_burned', 
             hue='fuel_type_code_pudl', 
             data=df[df.fuel_type_code_pudl != 'other']
            ).set_title("Avg Fuel Quantity Burned over time")
```

**Observation**: It shows that the qty of fuel increases for nuclear and gas.  Gas type qty has been increasing since year 2005 but the energy production has not been increased.  This shows that more gas is required to maintained the required amount of energy produced. It also shows that nuclear fule qty has been increasing since 2000 and more stablize after 2010.  There's also a spike showing between 1995 and 2000.

```python
energy_dict = {}
cols = df.fuel_type_code_pudl.unique().tolist()

for types in df.fuel_type_code_pudl.unique():
    energy_dict[types] = df[df.fuel_type_code_pudl == types]['energy_amount'].sum()
    
energy_amount_by_fuel_type = pd.DataFrame(energy_dict, index=[0])

sns.barplot(data=energy_amount_by_fuel_type, order=cols).set_title("Total energy produced by fuel type")
```

**Observation:**  Throughout the years we have coal produced the most energy.

```python
energy_dict = {}
cols = df.fuel_type_code_pudl.unique().tolist()

for types in df.fuel_type_code_pudl.unique():
    energy_dict[types] = df[df.fuel_type_code_pudl == types]['fuel_cost_per_unit_mmbtu'].sum()
    
fuel_cost_by_fuel_type = pd.DataFrame(energy_dict, index=[0])

sns.barplot(data=fuel_cost_by_fuel_type, order=cols).set_title("Total cost per unit mmbtu by fuel type")
```

**Observation:**  Comparatively, oil actually cost more than coal but produced relatively low energy if compared to coal.

```python
cost_efficiency = energy_amount_by_fuel_type/fuel_cost_by_fuel_type

sns.barplot(data=cost_efficiency, order=cols).set_title("Total cost efficiency by fuel type")
```

**Observation:**  However, for energy wise perspective, nuclear fuel is more cost efficient because it requires smallest amount of fuel cost to substantial amount of energy.  Coal and gas are likely to be less efficient when comparing to nuclear.

```python
sns.barplot(x='report_year',y='energy_amount',
            data=df_drop_outlier, hue='fuel_type_code_pudl',
            estimator=sum, order=[x for x in range(2014,2019)] 
           )
```

**Observation:**  We can see that the total amount of energy produce has been declining for coal in last five years.  This shows that the amount of energy needed by the country is reducing.  It could be a lot of factors for example the awareness of energy saving or implementation of LED and digital device that have low power consumption.

```python
sns.barplot(x='report_year',y='fuel_qty_burned',
            data=df, hue='fuel_type_code_pudl',
            estimator=sum, order=[x for x in range(2014,2019)] 
           )
```

**Observation:**  This graph shows that natural gas has been commonly use to power electricity for the country and has been increasing for the last five years.  It can be referring to major consumption of natural gas as a resources to produce thermal energy.  The high amount of resources consumption is obvious here when coal and nuclear is below 0.5e9 the last five years.

**Conclusion:** Coal and gas can be the major contribution to energy consumption in the country and also the major part of green house effect.  The good news is we show a decline in coal fuel to produce electricity.  However, we have been seeing an increase of natural gas usage for electricity to the coutnry.
